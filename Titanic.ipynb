```python
import pandas as pd
import numpy as np
```

```python
train= pd.read_csv(r"/content/train.csv")
test= pd.read_csv(r"/content/test.csv")
```

```python
train.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
modelo=RandomForestClassifier(n_estimators=100, n_jobs=-1, random_state=0)
```

```python
Variaveis=['Sex_binario','Age']
```

```python
def transformar(valor):
  if valor=='female':
    return 1
  else:
    return 0

train['Sex_binario']=train['Sex'].map(transformar)
```

```python
x=train[Variaveis]
y= train['Survived']
```

```python
x=x.fillna(-1)
```

```python
modelo.fit(x,y)
```

Agora irá aplicar o modelo nos dados de teste

```python
test['Sex_binario']=test['Sex'].map(transformar)
x_prev=test[Variaveis]
x_prev
```

```python
x_prev=x_prev.fillna(-1)
```

```python
p=modelo.predict(x_prev)
p
```

```python
sub=pd.Series(p, index=test['PassengerId'], name= "Survived" )
```

```python
sub.to_csv("Primeiro_modelo",header=True)
```
